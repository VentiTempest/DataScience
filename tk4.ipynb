```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import graphviz

df = pd.read_excel('datatk4.xlsx')

le = LabelEncoder()
df['JENIS KELAMIN'] = le.fit_transform(df['JENIS KELAMIN'])
df['STATUS MAHASISWA'] = le.fit_transform(df['STATUS MAHASISWA'])
df['STATUS PERNIKAHAN'] = le.fit_transform(df['STATUS PERNIKAHAN'])
df['STATUS KELULUSAN'] = le.fit_transform(df['STATUS KELULUSAN'])

X = df.drop(columns=['STATUS KELULUSAN'])
y = df['STATUS KELULUSAN']

clf = DecisionTreeClassifier()
clf = clf.fit(X, y)

dot_data = tree.export_graphviz(clf, out_file=None,
                                feature_names=['JENIS KELAMIN', 'STATUS MAHASISWA', 'STATUS PERNIKAHAN', 'IPK SEMESTER 1-6', 'Fitur_Tambahan'],  # Sesuaikan dengan nama fitur yang sebenarnya
                                class_names=['TEPAT', 'TERLAMBAT'],
                                filled=True, rounded=True,
                                special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("Decision_Tree")

graph

```
